```python
#Miramos la cantidad de valores nulos en la columna: 'Insulin'
Insulina = (data['Insulin'] == 0).sum()
Insulina
```

```python
#Cambiamos los valores nulos por la media de los demás.
valores_insulina = data['Insulin'].dropna()
promedio_insulina = sum(valores_insulina)/len(valores_insulina)
data['Insulin'] = data['Insulin'].replace(0, np.nan)
data['Insulin'] = data['Insulin'].fillna(promedio_insulina)
data['Insulin']
```

```python
#Miramos la cantidad de valores nulos en la columna: 'Glucose'
Glucosa = (data['Glucose'] == 0).sum()
Glucosa
```

```python
#Miramos la cantidad de valores nulos en la columna: 'BloodPressure'
Blood_Pressure = (data['BloodPressure'] == 0).sum()
Blood_Pressure
```

```python
#Miramos la cantidad de valores nulos en la columna: 'SkinThickness'
SkinThickness = (data['SkinThickness'] == 0).sum()
SkinThickness
```

```python
#Cambiamos los valores nulos por la media de los demás.
valores_SkinThickness = data['SkinThickness'].dropna()
promedio_SkinThickness = sum(valores_SkinThickness)/len(valores_SkinThickness)
data['SkinThickness'] = data['SkinThickness'].replace(0, np.nan)
data['SkinThickness'] = data['SkinThickness'].fillna(promedio_SkinThickness)
data['SkinThickness']
```

```python
#Miramos la cantidad de valores nulos en la columna: 'BMI'
BMI = (data['BMI'] == 0).sum()
BMI
```

```python
#Miramos la cantidad de valores nulos en la columna: 'DiabetesPedigreeFunction'
Diabetes_PedigreeFunction = (data['DiabetesPedigreeFunction'] == 0).sum()
Diabetes_PedigreeFunction
```

```python
#Miramos la cantidad de valores nulos en la columna: 'Pregnancies'
Pregnancies = (data['Pregnancies'] == 0).sum()
Pregnancies
```

```python
#Cambiamos los valores nulos por la media de los demás.
valores_Pregnancies = data['Pregnancies'].dropna()
promedio_Pregnancies = sum(valores_Pregnancies)/len(valores_Pregnancies)
data['Pregnancies'] = data['Pregnancies'].replace(0, np.nan)
data['Pregnancies'] = data['Pregnancies'].fillna(promedio_Pregnancies)
data['Pregnancies']
```

Lo que hicimos fue analizar si habían valores nulos, si la cantidad de estos es muy grande comparado con el total, lo cambiamos por el promedio de los demás datos de la columna en donde se esté haciendo el cambio, para ello hice un cambio de los valores nulos por valores de tipo (np.nan) para facilitar el intercambio en los datos.
